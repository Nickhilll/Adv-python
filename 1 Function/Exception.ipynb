## 1. Generic exception handling

```python
def division2():
    try:
        num = float(input('please enter numerator:'))
        den = float(input('Please enter denomenator:'))
        div = num/den
        print(f'Numerator {num}, Denominator {den} and Division is : {div:.4f}')
        return div
    except Exception as e:
        print(f'Exception Occured: {e}')
```

```python
# NUM - 12, DEN-5
d1 = division2()
print('Done')
```

```python
d2 = division2()
print('Done')
```

```python
d3 = division2()
print('Done')
```

# Error handling for different exceptions
1. Zero Division Error
2. value Error

```python
def division3():
    try:
        num = float(input('Please enter Numerator:'))
        den = float(input('Please enter Denominator:'))
        div = num/den
        print(f'Numerator: {num}, Denominator {den} and Division is : {div:.4f}')
        return div
    except ZeroDivisionError:
        print('Denominator in division cannot be zero')
    except ValueError:
        print('Numerator and Denominator should both be numbers, blank input is not allowed')
```

```python
# Num - 10, Den - 3
d1 = division3()
print("done")
```

```python
d1
```

```python
# Num - 20, Den - 0
d2 = division3()
print("done")
```

```python
# Num - d
d3 = division3()
print("done")
```

```python
# Num - blank
d3 = division3()
print("done")
```

##  Blocks in exception handling
1. try: the code which you try to execute
2. except: if any error or exception occurs then this block will get execute.
3. else: if the try block is executed only then else block will get executed
4. finally: Will always be executed for any block is executed


```python
from math import sqrt
```

```python
sqrt(576)
```

```python
sqrt(-2)
```

```python
def square_root():
    n = float(input('Enter a positive number:'))
    s = sqrt(n)
    print(f'SQuare root of given number {n} is: {s:.4f}')
    return s
```

```python
s1 = square_root()
print('Done')
```

```python
# n = "e"
s3 = square_root()
print("done")
```

```python
def square_root2():
    try:
        n = float(input('Enter a positive number:'))
        s = sqrt(n)
    except Exception as e:
        print(f'Exception occured :{e}')
    else:
        print(f'Square root of {n} is: {s:.4f}')
    finally:
        print('Finally Block is always executed')
        
```

```python
square_root2()
```

## raising custom error

```python
def age_validation():
    age = int(input('please enter age in years:'))
    if age <= 0 or age >= 100:
        raise ValueError('Age cannot less than 0 and greater than 100')
    else:
        print(f'Valid age entered: {age}')
        return age
```

```python
a1 = age_validation()
```

```python
a2 = age_validation()
```

```python
a3 = age_validation()
```

## File handling  

1. W -Write, create anew file and write the content
2. r - read, read the file and import data in python 
3. a - appent, add something in an existing file 

```python
# Save the content in new file using write file
with open('example.txt', 'w') as f1:
    f1.write('This is line1\n')
    f1.write('this is line2\n')
```

```python
# Read mode to read the context from the file
with open('example.txt','r') as f2:
    content  = f2.read()
```

```python
print(content)
```

```python
# Append mode to add new content in file
with open("example.txt", "a") as f3:
    f3.write("This is line 3\n")
    f3.write("Thank you\n")
```
