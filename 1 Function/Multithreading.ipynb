## Single thread processes
by default coding in python is single threaded

```python
import time
```

```python
%%time
time.sleep(2)
print('hello!')
```

```python
def hypotenuose(a, b):
    print('hypotenuose code execution started')
    time.sleep(2)
    h = (a**2+ b**2)**(1/2)
    print(f'hypotenuose of side {a} and {b} is {h:.4f}')
    print('Hypotenuose code completed')
```

```python
%%time 
hypotenuose(3,4)
```

```python
def simple_interest(p, n, r):
    print('Simple interest code execution started !')
    time.sleep(3)
    i = (p*n*r)/100
    a = p+i
    print(f'principle : {p} INR, Number of years: {n}, rate of interest {r} %p.a.')
    print(f'Interest: {i:.2f} INR')
    print(f'Amount: {a:.2f} INR')
    print('Simple Interest code execution done')
```

```python
%%time
simple_interest(p = 50_00_000, n = 6, r = 14.5)
```

```python
%%time 
hypotenuose(4, 5)# 2 seconds
simple_interest(p = 60_000_000, n = 6, r = 7.1) # 3 second
```

## Execute above functions simutaneously using multithreading

```python
from threading import Thread
```

```python
%%time
# Create the thread
th1 = Thread(target = hypotenuose, args = (12, 13))
th2 = Thread(target = simple_interest, args = (73,000, 6, 6.4))

# Start all the thread
th1.start()
th2.start()

# Wait for all threads to finish
th1.join()
th2.join()
```
